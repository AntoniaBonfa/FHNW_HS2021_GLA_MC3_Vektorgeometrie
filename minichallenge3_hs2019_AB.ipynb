## Aufgabe 1: Geraden im zweidimensionalen Raum

Geraden im zweidimensionalen Raum k√∂nnen auf unterschiedliche Arten beschrieben werden. Dabei unterscheidet man zwischen vektoriellen Beschreibungsformen und der Beschreibung mittels Koordinatengleichungen. In dieser √úbung geht es vorwiegend um die Beschreibung mittels Koordinatengleichungen. 

Hier finden Sie eine Zusammenstellung der wichtigsten Beschreibungsformen mittels Koordinatengleichung:

* Gerade als lineare Funktion ($m$ ist die Steigung und $b$ der $y$-Achsenabschnitt der Geraden:

$$y=m\cdot x + b$$

* Normalform der Geraden als Koordinatengleichung ($A$ und $B$ sind die $x$- und $y$-Komponenten des Normalenvektors, in $C$ ist die Information √ºber die Lage der Geraden versteckt):

$$A\cdot x+B\cdot y+C=0$$

* Normalform mit Vektoren ($\overrightarrow{n}$ ist der Normalenvektor (steht rechtwinklig zur Geraden) und $\overrightarrow{r_0}$ ist ein gegebener Punkt auf der Geraden):

$$\overrightarrow{n}\circ\left(\overrightarrow{r}-\overrightarrow{r_0}\right)=0$$

* Punkt-Steigungsform ($m$ ist die Steigung der Geraden und $P\left(x_p,y_p\right)$ ein Punkt auf der Geraden):

$$y = m\cdot\left(x-x_p\right)+y_p$$

* Zweipunktform ($P_1\left(x_1,y_1\right)$ und $P_2\left(x_2,y_2\right)$ sind zwei Punkte auf der Geraden):

$$y = \frac{y_2-y_1}{x_2-x_1}\cdot\left(x-x_1\right)+y_1$$

Wenn zwei Geraden im zweidimensionalen Raum mittels Koordinatengleichungen gegeben sind, z.B.

$$\begin{eqnarray*}g_1:&\;A_1 \cdot x + B_1 \cdot y + C_1 = 0\\g_2:&\;A_2 \cdot x + B_2 \cdot y + C_2 = 0\end{eqnarray*}$$

so ergibt sich der Schnittpunkt als L√∂sung des linearen Gleichungssystems:

$$\left|\begin{eqnarray}A_1 \cdot x + B_1 \cdot y + C_1 &=& 0\\ A_2 \cdot x + B_2 \cdot y + C_2 &=& 0\end{eqnarray}\right|$$

### Beispiel

Bestimmen Sie die Schnittpunkte der Geraden $g:\;y=2x+1$ und $h:\;x+4y+12=0$.

```python
from sympy import symbols
from sympy import solve
from sympy.plotting import plot
x = symbols('x')
y = symbols('y')
g_bsp = y-(2*x+1)
h_bsp = x+4*y+12
S = solve([g_bsp,h_bsp],x,y)
print('Schnittpunkt: ',S)
y_g = solve(g_bsp,y)
y_h = solve(h_bsp,y)
p=plot(y_g[0],y_h[0], show = False, legend = True)
p[1].line_color = 'r'
p.show()
```

## Aufgabe 3: Abstand eines Punktes zu einer Geraden (im zweidimensionalen Raum)

Wird die Normalengleichung der Geraden ($A\cdot x+B\cdot y+C=0$ bzw. $\overrightarrow{n}\circ\left(\overrightarrow{r}-\overrightarrow{r_0}\right)=0$) beidseitig durch die L√§nge des Normalenvektors ($\overrightarrow{n}=\left(\begin{array}{c}A\\B\end{array}\right)$) dividiert, erh√§lt man die Hesse'sche Normalform der Geraden:

$$\frac{A\cdot x+B\cdot y+C}{\sqrt{A^2+B^2}}=0$$

bzw.

$$\frac{\overrightarrow{n}\circ\left(\overrightarrow{r}-\overrightarrow{r_0}\right)}{\left|\overrightarrow{n}\right|}=0$$

Mittels der Hesse'schen Normalform haben wir die M√∂glichkeit den (k√ºrzesten) Abstand eines Punktes $Q\left(x_q,y_q\right)$ von einer Geraden zu bestimmen. Dazu wird der gegebene Punkt in die Hesse'schen Normalform eingesetzt:

$$\pm d=\frac{A\cdot x_q+B\cdot y_q+C}{\sqrt{A^2+B^2}}$$

bzw.

$$\pm d=\frac{\overrightarrow{n}\circ\left(\overrightarrow{r_Q}-\overrightarrow{r_0}\right)}{\left|\overrightarrow{n}\right|}$$

Erhalten wir beim Einsetzen des Punktes den Wert $0$, so liegt der Punkt auf der Geraden (sein Abstand zur Geraden ist Null!). Ist der Wert ungleich $0$, so liegt der Punkt nicht auf der Geraden. Dabei gibt der Betrag den Abstand an:

$$d=\left|\frac{A\cdot x_q+B\cdot y_q+C}{\sqrt{A^2+B^2}}\right|$$

bzw.

$$d=\left|\frac{\overrightarrow{n}\circ\left(\overrightarrow{r_Q}-\overrightarrow{r_0}\right)}{\left|\overrightarrow{n}\right|}\right|$$

Das Vorzeichen gibt an, auf welcher Seite der Geraden sich der Punkt $Q$ befindet (positiv: Punkt liegt auf der Seite, auf welche der Normalenvektor der Geraden hinzeigt / negativ: gerade umgekehrt!).



Die Distanz $d$ wird im Machine Learning zum Beispiel f√ºr [Support Vector Machines](https://en.wikipedia.org/wiki/Support-vector_machine) benutzt.

### Beispiel

Bestimmen Sie die Abst√§nde der Punkte $A\left(1,3\right)$, $B\left(3,1\right)$ und $C\left(2,2\right)$ zur Geraden $g:\; 5\cdot x-12\cdot y+52=0$

```python
from sympy import symbols
from sympy import evalf
import math
x = symbols('x')
y = symbols('y')
g = 5*x-12*y+52
n_x = 5
n_y = -12
abs_n = math.sqrt(n_x**2+n_y**2)
hnf_g = g/abs_n

A = [1,3]
B = [3,1]
C = [2,2]
d_A = hnf_g.evalf(subs={x:A[0],y:A[1]})
print('d(A,g) = ',d_A)
d_B = hnf_g.evalf(subs={x:B[0],y:B[1]})
print('d(B,g) = ',d_B)
d_C = hnf_g.evalf(subs={x:C[0],y:C[1]})
print('d(C,g) = ',d_C)
```

## Aufgabe 4: Spiegelung eines Punktes an einer Geraden (im zweidimensionalen Raum)

Der Punkt $Q$ soll an der Geraden $g:\;\overrightarrow{n}\circ\left(\overrightarrow{r}-\overrightarrow{r_0}\right)=0$ gespiegelt werden. Mittels der Hesse'schen Normalform k√∂nnen wir den (k√ºrzesten) Abstand bestimmen:

$$d=\left|\frac{\overrightarrow{n}\circ\left(\overrightarrow{r_Q}-\overrightarrow{r_0}\right)}{\left|\overrightarrow{n}\right|}\right|$$

Der gespiegelte Punkt $S$ hat von der Geraden $g$ den gleichen Abstand (einfach auf der anderen Seite der Geraden) und somit vom gegebenen Punkt $Q$ den Abstand $2d$.

Im weiteren gibt der Normalenvektor $\overrightarrow{n}$ der Geraden die Richtung des Verbindungsvektors $\overrightarrow{QS}$ an. Nun muss nur noch die L√§nge angepasst werden. F√ºr den gespiegelten Punkt findet man somit:

$$\overrightarrow{r_S}=\overrightarrow{r_Q}+\overrightarrow{QS}=\overrightarrow{r_Q}-\frac{2\frac{\overrightarrow{n}\circ\left(\overrightarrow{r_Q}-\overrightarrow{r_0}\right)}{\left|\overrightarrow{n}\right|}}{\left|\overrightarrow{n}\right|}\cdot\overrightarrow{n}=\overrightarrow{r_Q}-\frac{2\left(\overrightarrow{n}\circ\left(\overrightarrow{r_Q}-\overrightarrow{r_0}\right)\right)}{\left|\overrightarrow{n}\right|^2}\cdot\overrightarrow{n}$$

### Beispiel

Spiegeln Sie den Punkt $P\left(3,1\right)$ an der Geraden $g:\;x-y+1=0$.

```python
from sympy import symbols
from sympy import evalf
import math
from sympy import solve
import numpy as np

x, y = symbols(['x', 'y'])
g = x-y+1
P = np.array([3,1])
n = np.array([1,-1])
abs_n = np.linalg.norm(n)
hnf_g = g/abs_n

S = P - 2*hnf_g.evalf(subs={x:P[0], y:P[1]}) * n/abs_n
print('S = ',np.round(np.array(S, dtype=float),5))
```

## Aufgabe 5: Reflexion eines Strahls an einer Geraden (im zweidimensionalen Raum)

Beim Billiard / Snooker werden die Kugeln oft √ºber eine oder mehrere Banden gespielt (einem Hindernis ausweichen / der direkte Weg ist versperrt). Beim Spielen √ºber eine Bande gilt das Reflexionsgesetz (Einfallswinkel gleich Ausfallswinkel):


M√∂chte man nun eine Kugel (Position $A$) √ºber eine Bande auf eine zweite Kugel (Position $B$) spielen, so zielt man auf die (virtuell) gespiegelte Zielkugel (Position $B^{\prime}$):


### Teilaufgabe 5a)

Im Punkt $A\left(0,0\right)$ ist eine punktf√∂rmige Lichtquelle angebracht. In welche Richtung muss ein Lichtstrahl ausgesendet werden, so dass er nach Reflexion am Spiegel mit der Gleichung $g:\;3x+y+4=0$ durch den Punkt $B\left(2,5\right)$ geht? **Visualisieren Sie Ihre Resultate**.

```python
# IHR CODE
x,y = symbols(['x', 'y'])
A = [0,0]
B = [2,5]
i = 3*x + 1*y + 4

#Spiegelung A berechnen:
n = np.array([3,1])
abs_n = np.linalg.norm(n)
hnf = i/abs_n

Sp_A = A - 2*hnf.evalf(subs={x:A[0], y:A[1]}) * n/abs_n #Spiegelpunkt von A
Sp_B = B - 2*hnf.evalf(subs={x:B[0], y:B[1]}) * n/abs_n #Spiegelpunkt von B

A_SpB = ((Sp_B[1] - A[1]) / (Sp_B[0] - A[0])) * (x-A[0]) + A[1] -y #Gerade von A nach B'
B_SpA = ((Sp_A[1] - B[1]) / (Sp_A[0] - B[0])) * (x-B[0]) + B[1] -y #Gerade von A' nach B

S, = list(linsolve([A_SpB,B_SpA], x,y))
Sc = [S[0], S[1]]
print(Sc,Sp_A)

```
